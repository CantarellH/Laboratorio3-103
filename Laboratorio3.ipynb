# Obtener la Serie de Datos (*GRAFICA Vertical*)

```python
import pandas as pd
import matplotlib.pyplot as plt

path = "/content/drive/MyDrive/Conjunto de datos/atenciones_abogadas_de_las_mujeres (1).csv"

df = pd.read_csv (path)

edad = df["usuaria_edad"]
dist = edad.value_counts()
plt.bar(dist.index, dist.values)
plt.xlim(0, 100)
plt.show()
```
